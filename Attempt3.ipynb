## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df.drop(columns=['EIN', 'NAME'], inplace=True)
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()
application_type_counts
```

```python
# Create bins for Application type
num_bins_app_type = 6

application_type_bins= pd.qcut(application_df['APPLICATION_TYPE'].value_counts(),
                              q=num_bins_app_type, labels=[f'Bin_{i+1}' for i in range(num_bins_app_type)], duplicates='drop')
```

```python
# Create a dictionary to map application types to their bins
application_type_bins_dict = dict(zip(application_type_counts.index, application_type_bins))

# Map bins back to original df
application_df['APPLICATION_TYPE_BINS'] = application_df['APPLICATION_TYPE'].map(application_type_bins_dict)

# Check distribution of bins
print(application_df['APPLICATION_TYPE_BINS'].value_counts())
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
classification_counts = application_df['CLASSIFICATION'].value_counts()
classification_counts
```

```python
# Create bins for Classification
num_bins_classification = 6

classification_bins= pd.cut(application_df['CLASSIFICATION'].value_counts(),
                              bins=num_bins_classification, labels=[f'Bin_{i+1}' for i in range(num_bins_classification)], include_lowest=True)
```

```python
# Create a dictionary to map classifications to their bins
classification_bins_dict = dict(zip(classification_counts.index, classification_bins))

# Map bins back to original df
application_df['CLASSIFICATION_BINS'] = application_df['CLASSIFICATION'].map(classification_bins_dict)

# Check distribution of bins
print(application_df['CLASSIFICATION_BINS'].value_counts())
```

```python
# Drop original application_type and classification columns
application_df.drop(columns=['APPLICATION_TYPE', 'CLASSIFICATION'], inplace=True)
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df_numeric = pd.get_dummies(application_df)
application_df_numeric.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df_numeric['IS_SUCCESSFUL']
X = application_df_numeric.drop(columns='IS_SUCCESSFUL')

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
input_features = len(X_train_scaled[0])
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=8, activation="relu", input_dim=input_features))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=8, activation="relu"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=8, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=50)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
